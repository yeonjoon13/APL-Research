```python
#Branch et al. 2018
from intern import array
import matplotlib.pyplot as plt
```

```python
#importing mouse brain connectome - resolution 0
mouse_res_0 = array("bossdb://ABagemouse/dentateECcagGFP3/Ch0")

#shows size of original dataset
mouse_res_0.shape
```

```python
#accessing a slice of data from Branch dataset
mouse_slice_0 = mouse_res_0[250:255, 0:2560, 0:2160]

#Show the plot of slice
plt.imshow(mouse_slice_0[0])
```

```python
#check to see if data has been downsampled
print(mouse_res_0._channel.downsample_status)
```

```python
#check to see #s of levels of downsampling
print(mouse_res_0._exp.num_hierarchy_levels, "levels of downsampling")
```

```python
#4 levels of downsampling, has been downsampled 3 times

#lower resolution - res 3
mouse_res_3 = array("bossdb://ABagemouse/dentateECcagGFP3/Ch0", resolution = 3)

#size of dataset
print(mouse_res_3.shape)

#accessing a slice of data from Branch dataset
mouse_slice_3 = mouse_res_3[250:255, 0:320, 0:270]

#show plot of slice
plt.imshow(mouse_slice_3[0])
```

```python
#even lower resolution - res 1
mouse_res_1 = array("bossdb://ABagemouse/dentateECcagGFP3/Ch0", resolution = 1)

#size of dataset
print(mouse_res_1.shape)

#accessing a slice of data from Branch dataset
mouse_slice_1 = mouse_res_1[250:255, 0:1280, 0:1080]

#show plot of slice
plt.imshow(mouse_slice_1[0])
```

```python
mouse_slice_0 = mouse_res_0[250:255, 0:2560, 0:2160]
mouse_slice_3 = mouse_res_3[250:255, 0:320, 0:270]
mouse_slice_1 = mouse_res_1[250:255, 0:1280, 0:1080]

#subplots of data
fig, axes = plt.subplots(3, 1, figsize=(40,10))
axes[0].imshow(mouse_slice_0[0])
axes[1].imshow(mouse_slice_3[0])
axes[2].imshow(mouse_slice_1[0])
plt.show()
```
